```python
pair_plot = df[['bedrooms', 'sqft_living', 'price','grade']]             #show pair-plot result  (scatter plot)
sns.pairplot(pair_plot)
```

```python
sum = df.groupby(by='grade').sum()
mean = df.groupby(by='grade').mean()
count = df.groupby(by='grade').count()
```

```python
plt.title('Mean Price by Grade')
plt.xlabel('Grade') 
plt.ylabel('Mean Price')
plt.bar(mean.index, mean.price)
plt.grid()
plt.show()
```

```python
df2 = df.groupby(by='zipcode').mean()

plt.figure(figsize=(14,30))
df2['price'].sort_values().plot(kind='barh', grid=True)
plt.title('House Price by Zipcode')
plt.xlabel('Mean Price')
plt.ylabel('Zipcode') 
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.scatter(df.sqft_living, df.price)
slope, intercept = np.polyfit(df.sqft_living, df.price, 1)
plt.plot(df.sqft_living, df.sqft_living * slope + intercept, "r")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.scatter(df2.sqft_living, df2.price)
slope, intercept = np.polyfit(df2.sqft_living, df2.price, 1)
plt.plot(df2.sqft_living, df2.sqft_living * slope + intercept, "r")
plt.show()
```

```python
df2['offset'] = df2['price'] - (df2.sqft_living*slope + intercept)
df2['offset2'] = np.abs(df2['price'] - (df2.sqft_living*slope + intercept))
df2_sort = df2.sort_values(by = 'offset2', ascending=False)
df2_sort.head()
```
