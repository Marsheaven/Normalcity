```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('test.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(12,6))
plt.scatter(df.sqft_living, df.price)
slope, intercept = np.polyfit(df.sqft_living, df.price, 1)
plt.plot(df.sqft_living, (df.sqft_living * slope + intercept), "r")
for n in range(3):
    plt.text(df['sqft_living'][n], df['price'][n], df['zipcode'][n])

for n in range(67,70,1):
    plt.text(df['sqft_living'][n], df['price'][n], df['zipcode'][n])
    
#plt.text(df['sqft_living'][69], df['price'][69], df['zipcode'][69])
plt.show()
```
