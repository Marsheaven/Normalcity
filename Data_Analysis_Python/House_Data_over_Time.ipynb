```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('C:\\Users\\pc\\Documents\\Python_2019\house_data.csv')
df.head(3)
```

```python
df2 = df.copy()
df2['date'] = pd.to_datetime(df['date']).dt.strftime('%Y-%m')
df2.head(3)
```

```python
df2_mean = df2.groupby(by='date').mean()
df2_mean.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(15,10))
plt.plot(df2_mean.index, df2_mean.price)
plt.xlabel = ('Date')
plt.ylabel = ('Mean Price')
plt.title('House mean price (2014 - 2015)')
plt.grid()
```

```python
df_98029= df2[df2.zipcode == 98029]
df_98029 = df_98029.sort_values(by='date', ascending=True)
df_98029.head()
```

```python
df_98112= df2[df2.zipcode == 98112]
df_98112 = df_98112.sort_values(by='date', ascending=True)
df_98112.head()
```

```python
df_98029_mean = df_98029.groupby(by='date').mean()
df_98112_mean = df_98112.groupby(by='date').mean()
```

```python
df_98029_mean.shape, df_98112_mean.shape
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(15,10))
plt.plot(df_98029_mean.index, df_98029_mean.price, label = '98029')
plt.plot(df_98029_mean.index, df_98112_mean.price, label = '98112')
plt.xlabel = ('Date')
plt.ylabel = ('Price')
plt.title('House mean price over time')
plt.legend()
plt.grid()
plt.show()
```
