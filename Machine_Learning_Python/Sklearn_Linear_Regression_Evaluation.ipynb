```python
# retrying the file reading by separating columns
df2 = pd.read_csv('C:\\Users\\pc\\Documents\\Kaggle\housing_Boston.csv', header=None, sep='\s+')
# \s+ means one or more whitespace characters.
# ' ' for a space or '\t' for a tab
df2.head()
```

```python
df2.shape
```

```python
# total number of the columns is 14
# assigning the column names
df2.columns = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
df2.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

cols = ['CRIM','INDUS','NOX','RM', 'LSTAT', 'MEDV']
sns.pairplot(df2[cols], height=2.5)
#plt.tight_layout()
plt.show()
```

```python
corr = round(df2.corr(method='pearson'), 2)
corr
```

```python
# sklearn (linearRegression) modeling module import
from sklearn.linear_model import LinearRegression
# train/test data split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df2.drop(['MEDV'], axis=1), df2['MEDV'], test_size = 0.3, random_state = 5)
```

```python
# fitting the model to train the data
lr = LinearRegression()
model = lr.fit(x_train, y_train)
```

```python
# prediction by test data
pred_y = lr.predict(x_test)
```

```python
# compare the predicted result(pred_y) to the actural data(y_test)
result = pd.DataFrame({'Actual' : y_test, 'Predicted': pred_y})
result.head()
```

```python
result2 = result.head(15)
result2.plot(kind='bar',figsize=(10,8))
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.show()
```

```python
x_test['PRED']=pred_y
x_test['MEDV']=y_test
x_test.head()
```

```python
# Model Evaluation (MSE, RMSE)
import sklearn
from sklearn.metrics import mean_squared_error
mae = sklearn.metrics.mean_absolute_error(y_test, pred_y)
mse = sklearn.metrics.mean_squared_error(y_test, pred_y)

print('MAE(Mean Absolute Error):', mae)  
print('MSE(Mean Squared Error):', mse) 
print('RMSE(Root Mean Squared Error):', np.sqrt(mean_squared_error(y_test, pred_y)))
```

```python
# MSE for train data & test data

pred_y_train = lr.predict(x_train)
mse_test = sklearn.metrics.mean_squared_error(y_test, pred_y)
mse_train = sklearn.metrics.mean_squared_error(y_train, pred_y_train)
print("train MSE : %.3f, test MSE : %.3f" % (mse_train, mse_test))
```

```python
# R2(R-square) score for train and test data
import sklearn
from sklearn.metrics import r2_score

print("train R2 : %.3f, test R2 : %.3f" % (r2_score(y_train, pred_y_train), r2_score(y_test, pred_y)))
```

```python
# Offset/Residual graph analysis

#pred_y = lr.predict(x_test)
#pred_y_train = lr.predict(x_train)

plt.scatter(pred_y_train, pred_y_train - y_train, c='steelblue', marker='o', edgecolors='white', label='Training Data')
plt.scatter(pred_y, pred_y - y_test, c='limegreen', marker='s', edgecolors='white', label='Test Data')
plt.xlabel("Predicted Values")
plt.ylabel("Residual")
plt.legend(loc="upper left")
plt.hlines(y=0, xmin=-10, xmax=50, color='black', lw=2)
plt.xlim([-10, 50])
#plt.tight_layout()
plt.show()
```
